```python
from pathlib import Path
import pandas as pd
from collections import defaultdict
```

```python
paczia_folder = Path('')
pdfs = []
measurements = defaultdict(list)
for pdf_fn in paczia_folder.glob('*.csv'):
    dfi = pd.read_csv(pdf_fn, index_col=0).iloc[1:]
    dfi.fillna(0, inplace=True)
    # Find first non-zero entry in each column
    first_nonzero_value = {}
    for col in dfi.columns:
        first_nonzero_value[col] = dfi.loc[dfi[col] != 0, col].values[0]
    
    # Divide each column by its first non-zero entry
    for col in dfi.columns:
        dfi[col] = dfi[col] / first_nonzero_value[col]
    for col in dfi.columns:
        if (dfi[col] > 2).sum() > 2:
            measurements[col].append(1)
        else:
            measurements[col].append(0)
            
```

```python
include = []
for met, vals in measurements.items():
    if (sum(vals) >= 3):
        include.append(met)
    else:
        print(f'skipping {met}, only {vals}')
```

```python
for x in sorted(include):
    print(x)
```
